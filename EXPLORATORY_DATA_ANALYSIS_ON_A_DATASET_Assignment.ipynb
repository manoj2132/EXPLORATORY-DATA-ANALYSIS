```python
import pandas as pd

# Assuming df is your numpy array
df = pd.DataFrame(df, columns=['LB','AC','FM','UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'Width', 'Tendency', 'NSP'])  # Add all your column names here

# Now you can check for missing values
print(df.isnull().sum())
```

```python
df.head(20)
```

```python
df.tail(15)
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR=Q3-Q1
outliers = (df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))
outliers_values = df[outliers]

outliers_values
```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))  # Set the figure size
sns.boxplot(data=df)
plt.title('Box Plot')  # Set the title
plt.show()
```

```python
df_no_outliers = df[~outliers.any(axis=1)]# this is use to remove of outliers
df_no_outliers
```

```python

plt.figure(figsize=(10, 10))
sns.boxplot(data=df_no_outliers)
plt.title('Boxplot')
plt.show()
```

In above plot have no outliers.

```python
df.hist(bins=20, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Variables')
plt.show()
```

```python
# There have no categorical value.so we do not Use bar charts or pie charts to display the frequency of categories for categorical variables.
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='FM', y='AC', data=df)
plt.title('Scatter Plot of var1 vs var2')
plt.xlabel('FM')
plt.ylabel('AC')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df, diag_kind='kde')
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

```python
#this pair plot is a powerful visualization that helps you quickly understand the relationships between multiple variables in your dataset.
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='AC', y='FM', data=df, palette='Set3')
plt.title('Violin Plot of num_var by category_column')
plt.show()
```
